```python
import numpy as np
import matplotlib.pyplot as plt
import torch
```

```python
raw_data = np.load('./pca_data_norm.npy')
```

```python
raw_data.shape
```

```python
fig1, ax1 = plt.subplots(figsize=(116,8))
xx = torch.linspace(0,11539,11539)
ax1.plot(xx, raw_data[0:11539,0], color='blue', label='x')
```

```python
# 下面进行采样，每20个点采样一个点
# 采样的目的是为了减少数据量，同时保留数据的特征
# 采样的方法是每隔20个点采样一个点

sample_interval = 20
sample_data = raw_data[::sample_interval]
sample_data.shape
```

```python
fig2, ax2 = plt.subplots(figsize=(116,8))
xx2 = torch.linspace(0,577,577)
ax2.plot(xx2, sample_data[:,0], color='blue', label='x')
```

```python
pca_train_downsample_20 = sample_data[:int(0.6*len(sample_data))]
pca_test_downsample_20 = sample_data[int(0.6*len(sample_data)):]
```

```python
print(pca_train_downsample_20.shape)
print(pca_test_downsample_20.shape)
```

```python
np.save('pca_train_downsample_20.npy', pca_train_downsample_20)
np.save('pca_test_downsample_20.npy', pca_test_downsample_20)
np.save('pca_downsample_20.npy', sample_data)
```
